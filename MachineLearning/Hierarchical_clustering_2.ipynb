```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import KMeans
```

```python
veriler  = pd.read_csv("musteriler.csv")

hacim_maas= veriler.iloc[:, 2:4].values
```

```python
agc= AgglomerativeClustering(n_clusters=4, affinity="euclidean", linkage="ward")
#affinity manhattan veya cosine olabilir
pred_agc= agc.fit_predict(hacim_maas)
```

```python
plt.scatter(hacim_maas[pred_agc==0,0], hacim_maas[pred_agc==0,1],s=100, c="red")
plt.scatter(hacim_maas[pred_agc==1,0], hacim_maas[pred_agc==1,1],s=100, c="green")
plt.scatter(hacim_maas[pred_agc==2,0], hacim_maas[pred_agc==2,1],s=100, c="blue")
plt.scatter(hacim_maas[pred_agc==3,0], hacim_maas[pred_agc==3,1],s=100, c="yellow")
plt.show()
```

```python
import scipy.cluster.hierarchy as sch

dendrogram= sch.dendrogram(sch.linkage(hacim_maas, method="ward"))
plt.show()
```

# kmeans and Agglomerative Clustering

```python

kmeans= KMeans(n_clusters=4, init="k-means++") #Random_state verilebilir.

kmeans.fit(hacim_maas)
pred_kmeans= kmeans.predict(hacim_maas)
print(kmeans.cluster_centers_)

sonuclar=[]

for i in range(1,10):
    kmeans= KMeans(n_clusters=i,init="k-means++", random_state=1)
    kmeans.fit(hacim_maas)
    sonuclar.append(kmeans.inertia_)

plt.plot(range(1,10), sonuclar)
plt.show()


plt.scatter(hacim_maas[pred_kmeans==0,0], hacim_maas[pred_kmeans==0,1],s=100, c="red")
plt.scatter(hacim_maas[pred_kmeans==1,0], hacim_maas[pred_kmeans==1,1],s=100, c="green")
plt.scatter(hacim_maas[pred_kmeans==2,0], hacim_maas[pred_kmeans==2,1],s=100, c="blue")
plt.scatter(hacim_maas[pred_kmeans==3,0], hacim_maas[pred_kmeans==3,1],s=100, c="yellow")
plt.show()
```
