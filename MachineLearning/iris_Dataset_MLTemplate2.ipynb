```python
# -*- coding: utf-8 -*-
"""
Created on Thu Jun 24 21:47:58 2021

@author: mucahitrtn
"""

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import confusion_matrix

iris= pd.read_excel("iris.xls")

from sklearn.model_selection import train_test_split

x = iris.iloc[:,0:4].values #bağımsız değişkenler
y = iris.iloc[:,4:].values #bağımlı değişken

x_train, x_test,y_train,y_test = train_test_split(x,y,test_size=0.33, random_state=0)
```

```python
#verilerin olceklenmesi
from sklearn.preprocessing import StandardScaler

sc=StandardScaler()

X_train = sc.fit_transform(x_train)
X_test = sc.transform(x_test)
```

```python

from sklearn.linear_model import LogisticRegression

logr= LogisticRegression(random_state=0)
logr.fit(X_train, y_train)

logr_pred= logr.predict(X_test)

# confusion matrix logr

logr_cfm= confusion_matrix(y_test, logr_pred)
```

```python
from sklearn.metrics import plot_confusion_matrix
print(logr_cfm)
plot_confusion_matrix(logr, x_test, y_test)
plt.show()
```

```python
# KNeighbors Classifier

from sklearn.neighbors import KNeighborsClassifier

knn= KNeighborsClassifier(n_neighbors=2, metric="minkowski")
knn.fit(X_train, y_train)

knn_pred= knn.predict(X_test)

knn_cfm= confusion_matrix(y_test, knn_pred)

print(knn_cfm)

plot_confusion_matrix(knn, x_test, y_test)
plt.show()
```

```python

# SUPPORT VECTOR MACHINE

from sklearn.svm import SVC

svc= SVC(kernel="rbf")
svc.fit(X_train, y_train)

svc_pred= svc.predict(X_test)

svc_cfm= confusion_matrix(y_test, svc_pred)
print(svc_cfm)
plot_confusion_matrix(svc, x_test, y_test)
plt.show()
```

```python
from sklearn.naive_bayes import GaussianNB

gnb= GaussianNB()

gnb.fit(X_train, y_train)

gnb_pred= gnb.predict(X_test)

gnb_cfm= confusion_matrix(y_test,gnb_pred)
print(gnb_cfm)
plot_confusion_matrix(gnb, x_test, y_test)
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier


dtc= DecisionTreeClassifier(criterion="entropy")
dtc.fit(X_train, y_train)

dtc_pred= dtc.predict(X_test)

dtc_cfm= confusion_matrix(y_test, dtc_pred)
print(dtc_cfm)
plot_confusion_matrix(dtc, x_test, y_test)
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier

rfc= RandomForestClassifier(n_estimators=10, criterion="entropy")

rfc.fit(X_train, y_train)

rfc_pred= rfc.predict(X_test)
rfc_cfm= confusion_matrix(y_test, rfc_pred)
print(rfc_cfm)
plot_confusion_matrix(rfc, x_test, y_test)
plt.show()
```
